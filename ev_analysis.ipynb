```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path="Electric_Vehicle_Population_Data.csv"
```

```python
df = pd.read_csv(file_path, low_memory=False)
```

```python
print("✅ File loaded successfully!")
```

```python
duplicates=df.duplicated().sum()
print(f"Duplicated rows found :{duplicates}")
if duplicates>0:
    df=df.drop_duplicates()
    print(f"Removed all the duplicated values")
```

```python
df.columns=df.columns.str.strip().str.lower().str.replace(" ","_")
df.columns
```

```python
missing_percent = df.isnull().mean() * 100
print(missing_percent.sort_values(ascending=False))
```

```python
threshold = 0.5
df = df.loc[:, df.isnull().mean() < threshold]
print(f"Columns retained: {len(df.columns)}")
```

```python
df = df.dropna(subset=['make', 'model', 'model_year'])
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna("NA")
```

```python
df['model_year'] = pd.to_numeric(df['model_year'], errors='coerce').astype('Int64')
```

```python
df['postal_code'] = df['postal_code'].astype(str)
```

```python
df['base_msrp'] = pd.to_numeric(df['base_msrp'], errors='coerce')
df['electric_range'] = pd.to_numeric(df['electric_range'], errors='coerce')
```

```python
print("✅ Final Cleaned Dataset Summary:")
print(df.info())
print(df.isnull().sum())
print(f"Final Shape: {df.shape}")
```

```python
#Which car brands leads the EV market?
top_makes=df['make'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_makes.index, y=top_makes.values, palette='viridis')
plt.title("Top 10 EV Manufacturers by Count")
plt.xlabel("Car Make")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=45)
plt.show()
```

```python
#Regions or cities which have the most EV registrations ?
city_counts = df['city'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(y=city_counts.index, x=city_counts.values, palette='magma')
plt.title("Top 10 cities with Most EVs")
plt.xlabel("Number of EVs")
plt.ylabel("City")
plt.show()
```

```python
#EVs Adoption trend by year?
yearly_trend = df['model_year'].value_counts().sort_index()
plt.figure(figsize=(10,5))
sns.lineplot(x=yearly_trend.index, y=yearly_trend.values, marker='o')
plt.title("EV Registration Trend Over the Years")
plt.xlabel("Model Year")
plt.ylabel("Number of EVs")
plt.show()
```

```python
#Average electric range across different car makers?
plt.figure(figsize=(12,6))
top_makes_range = df[df['make'].isin(df['make'].value_counts().head(5).index)]
sns.boxplot(x='make', y='electric_range', data=top_makes_range, palette='coolwarm')
plt.title("Electric Range Distribution by Top 5 Car Makes")
plt.xlabel("Car Make")
plt.ylabel("Electric Range (miles)")
plt.xticks(rotation=45)
plt.show()
```

```python
#Correlation between the vehicle’s price and its range?
plt.figure(figsize=(8,6))
sns.scatterplot(x='base_msrp', y='electric_range', data=df, alpha=0.6)
plt.title("Price vs Electric Range")
plt.xlabel("Base MSRP ($)")
plt.ylabel("Electric Range (miles)")
plt.show()
```

```python
#What is the distribution between battery-only and hybrid EVs?
if 'electric_vehicle_type' in df.columns:
    ev_type_counts = df['electric_vehicle_type'].value_counts()
    plt.figure(figsize=(6,6))
    plt.pie(ev_type_counts, labels=ev_type_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
    plt.title("Distribution of EV Types")
    plt.show()
```

```python
#Which EVs offers the best range ?
top_models = df.groupby('model')['electric_range'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_models.values, y=top_models.index, palette='crest')
plt.title("Top 10 EV Models by Average Electric Range")
plt.xlabel("Average Electric Range (miles)")
plt.ylabel("Model")
plt.show()
```

```python
#Which counties are adopting EVs faster ?
county_counts = df['county'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=county_counts.values, y=county_counts.index, palette='flare')
plt.title("Top 10 Counties by EV Population")
plt.xlabel("Number of EVs")
plt.ylabel("County")
plt.show()
```

```python
#relation between numeric valiables?
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Numeric Features")
plt.show()
```
